```python
import pandas as pd
import requests , bs4
targetUrl = "https://www.timeanddate.com/holidays/south-korea/"
resp = requests.get(url = targetUrl)
htmlObj = resp.text
bsObj = bs4.BeautifulSoup(htmlObj , "html.parser")
tableTag = bsObj.find(name = "table" , attrs = {"id" : "holidays-table"})
tbodyTag = tableTag.find(name = "tbody")
trTag = []
dateList= []
weekinfoList = []
holidayList = []
typeList = []
for i in range(4,36):
    targetTr = f"tr{i}"
    trTag.append(tbodyTag.find(name = "tr" , attrs = {"id" : targetTr}))
for j in range (0, len(trTag)):
    if trTag[j] is not None:
        thTag = trTag[j].find(name = "th")
        date = thTag.text
        dateList.append(date)
        tdTag = trTag[j].find(name = "td" , attrs = {"class" : "nw"})
        weekinfo = tdTag.text
        weekinfoList.append(weekinfo)
        tdTag = trTag[j].find(name = "a" )
        holiday_name = tdTag.text
        holidayList.append(holiday_name)
        tdTag = trTag[j].find_all(name ="td")
        tbType = tdTag[2].text
        typeList.append(tbType)
```

```python
df = pd.DataFrame(zip(dateList, weekinfoList,holidayList,typeList) , columns = ("date" , "weekinfo" , "holiday_name" , "type"))
```
