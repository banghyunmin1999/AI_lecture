```python
def preWeek(yearsWeek, targetvalue):
  """
  타겟 년도와 주차 그리고 줄이고 싶은 주차를 넣으면
  계산해 주는 함수
  yearsWeek -> int,str
  targetvalue -> int,str
  """
  #받은 주차-줄일주차
  weekTarget = 0

  #분리 하기 위해 타겟 년도와 주차 str화
  yearsWeek = str(yearsWeek)

  #int,str 어떤 형식으로 와도 int형식으로 줄이고 싶은 주차를 변경
  targetvalue = int(targetvalue)

  #타겟 년도와 주차 분리
  yearTarget = yearsWeek[0:4]
  weekTarget = yearsWeek[4:6]

  #계산하기 위해 타겟 년도와 주차 int 화
  yearTarget = int(yearTarget)
  weekTarget = int(weekTarget)

  #계산 시작
  while True :
    #타겟 주차 > 줄일 주차
    if(weekTarget > targetvalue):
        weekTarget = weekTarget - targetvalue
        break
    #타겟 주차 == 줄일 주차
    elif(weekTarget == targetvalue):
        yearTarget = yearTarget - 1 #타겟 년도-1 하고 타겟 주차를 이전년도 최고 주차로 변경
        weekTarget = Week.last_week_of_year(yearTarget).week
        break
    #타겟 주차 < 줄일 주차
    else:
      targetvalue = targetvalue - weekTarget
      yearTarget = yearTarget - 1 #타겟 년도-1 하고 타겟 주차를 이전년도 최고 주차로 변경
      weekTarget = Week.last_week_of_year(yearTarget).week

  #완성된 타겟 년도와 타겟 주차를 붙여서 출력 하기 위해 str 화
  yearTarget = str(yearTarget)
  weekTarget = str(weekTarget)

  #완성한것을 반환
  resultValue = yearTarget + weekTarget
  return(resultValue)
```
