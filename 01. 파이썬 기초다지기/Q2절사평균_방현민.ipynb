```python
import pandas as pd
```

```python
from binascii import b2a_base64
b2a_base64#깃허브의 www.github.com/hyokwakn -> python_lecture 클릭 dataset ->
#kopo_customerdata.csv
#https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv
csFilePath = "https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/kopo_customerdata.csv"
csData = pd.read_csv(csFilePath)
```

```python
customerTotalList = list( csData.TOTAL_AMOUNT)
customerTotalList.sort()

#최솟값 구하고 가장 오른쪽에 있는 최솟값의 인덱스를 구함
minCustomerTotalList = min(customerTotalList)
minCountCustomerTotalList = customerTotalList.count(minCustomerTotalList)
CountMinData = customerTotalList.count(minCustomerTotalList)

#최댓값 구하고 가장 오른쪽에 있는 최댓값의 인덱스를 구함
maxCustomerTotalList = max(customerTotalList)
CountMaxData = customerTotalList.count(maxCustomerTotalList)
maxCountCustomerTotalList = customerTotalList.count(maxCustomerTotalList)

#리스트 구간 삭제 [최댓값:] [:최솟값]
customerTotalList[len(customerTotalList)-CountMaxData:] = []
customerTotalList[:CountMinData] = []

avgCustomerTotalList = sum(customerTotalList) / len(customerTotalList)
print(int(avgCustomerTotalList))
```
